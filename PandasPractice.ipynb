* Pandas is an open source data analysis

*   pandas has two types of data structure:

1. Series - "It's a one dimensional array with indexes, it stores a single  column or row of data in a dataframe.
2. Dataframe - It's a tablular spreadsheet like structure representing rows each of which conatins one or multiple columns

* A one dimensional array (labeled) capable of holding any type of data - Series

*  A two dimensional data (labeled) structure with columns of potentially diffrent types of data - Dataframes.
